### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
#random.normal(loc=0.0, scale=1.0, size=None)
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st


normal_dist = np.random.normal(loc=0.0, scale=1.0, size=100)
plt.hist(normal_dist,bins=30)
plt.title('Normal Distribution')
plt.show
````

````python
#random.chisquare(df, size=None)
#- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
chi_square_distribution= np.random.chisquare(df=3, size=100)
plt.hist(chi_square_distribution,bins=30)
plt.title('Chi-Square Distribution')
plt.show
````

````python
#- Calculate the main metrics and statistical measures that best describe the two vectors.
# mean, median, mode, percentiles, range, variance, and standard deviation
normal_dist = np.random.normal(loc=0.0, scale=1.0, size=100)
chi_square_distribution= np.random.chisquare(df=3, size=100)

normal_mean = np.mean(normal_dist)
normal_median = np.median(normal_dist)
normal_mode = st.mode(normal_dist)
normal_percent_25 = np.percentile(normal_dist, 25)
normal_percent_50 = np.percentile(normal_dist, 50)
normal_percent_75 = np.percentile(normal_dist, 75)

plt.hist(normal_dist,bins=30, alpha=0.4, label = 'Normal Distribution')
plt.axvline(normal_mean, color ='green', alpha=0.6 ,label ='Median Mean Distribution')
plt.axvline(normal_median, color ='blue', alpha=0.6 ,label ='Median Normal Distribution')
#plt.axvline(normal_mode, color ='orange', alpha=0.6)
plt.axvline(normal_percent_25, color ='red', alpha=0.6,label='Percentil 25')
plt.axvline(normal_percent_50, color ='brown', alpha=0.6, label ='Percentil 50')
plt.axvline(normal_percent_75, color ='orange', alpha=0.6, label ='Percentil 75')
plt.legend(fontsize = 10) 
plt.show

````

````python
chi_square_mean = np.mean(chi_square_distribution)
chi_square_median = np.median(normal_dist)
chi_square_mode = st.mode(normal_dist)
chi_square_percent_25 = np.percentile(normal_dist, 25)
chi_square_percent_50 = np.percentile(normal_dist, 50)
chi_square_percent_75 = np.percentile(normal_dist, 75)

plt.hist(chi_square_distribution,bins=30, alpha=0.5, label = 'Chi-square distribution')
plt.axvline(chi_square_mean,color ='green', alpha=0.6 , label='Median Chi Square')
plt.axvline(chi_square_median,color ='blue', alpha=0.6 , label='Median Chi Square')
#plt.hist(chi_square_mode,bins=30, color ='orange', alpha=0.6)
plt.axvline(chi_square_percent_25,color ='red', alpha=0.6, label='Chi-square Percentil 25')
plt.axvline(chi_square_percent_50,color ='brown', alpha=0.6, label='Chi-square Percentil 50')
plt.axvline(chi_square_percent_75,color ='orange', alpha=0.6, label='Chi-square Percentil 75')
plt.legend(fontsize = 10) 
plt.show
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python

data = [4, 2, 5, 8, 6]
sd = np.std(data)
result = float(sd)
plt.hist(result,bins=30, alpha=0.8, color ='green')
plt.title('Standard Deviation')
plt.show
````
